```python
import turtle
pen = turtle.Turtle()

def draw_rectangle(cd,cr):
    pen.forward(cr)
    pen.left(90)
    pen.forward(cd)
    pen.left(90)
    pen.forward(cr)
    pen.left(90)
    pen.forward(cd)
    
pen.pencolor('SteelBlue2')
pen.penup()
pen.backward(150)
pen.pendown()
pen.fillcolor("IndianRed2")
pen.begin_fill()
draw_rectangle(200,150)
pen.end_fill()
pen.left(90)
pen.penup()
pen.forward(60)
pen.left(90)
pen.forward(60)
pen.right(90)
pen.pendown()
pen.fillcolor("gray100")
pen.begin_fill()
draw_rectangle(80,30)
pen.end_fill()
pen.penup()
pen.forward(60)
pen.left(90)
pen.forward(90)
pen.pendown()
pen.fillcolor("IndianRed2")
pen.begin_fill()
draw_rectangle(200,150)
pen.end_fill()
pen.left(90)
pen.penup()
pen.forward(60)
pen.left(90)
pen.forward(60)
pen.right(90)
pen.pendown()
pen.fillcolor("gray100")
pen.begin_fill()
draw_rectangle(80,30)
pen.end_fill()
pen.left(90)
pen.penup()
pen.goto(0,0)
pen.backward(150)
pen.right(90)
pen.forward(200)
pen.left(90)
pen.pendown()
pen.fillcolor("IndianRed2")
pen.begin_fill()
draw_rectangle(200,150)
pen.end_fill()
pen.left(90)
pen.penup()
pen.forward(60)
pen.left(90)
pen.forward(60)
pen.right(90)
pen.pendown()
pen.fillcolor("gray100")
pen.begin_fill()
draw_rectangle(80,30)
pen.end_fill()
pen.penup()
pen.forward(60)
pen.left(90)
pen.forward(90)
pen.pendown()
pen.fillcolor("IndianRed2")
pen.begin_fill()
draw_rectangle(200,150)
pen.end_fill()
pen.left(90)
pen.penup()
pen.forward(60)
pen.left(90)
pen.forward(60)
pen.right(90)
pen.pendown()
pen.fillcolor("gray100")
pen.begin_fill()
draw_rectangle(80,30)
pen.end_fill()
pen.penup()
pen.forward(60)
pen.left(90)
pen.forward(90)
pen.pendown()
pen.fillcolor("LemonChiffon4")
pen.begin_fill()
draw_rectangle(200,200)
pen.end_fill()
pen.left(90)
pen.forward(50)
pen.fillcolor("DarkSlateGray4")
pen.begin_fill()
draw_rectangle(100,100)
pen.end_fill()
pen.left(90)
pen.forward(50)
pen.left(90)
pen.forward(100)
```
